```python
import pandas as pd
import numpy as np
```

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
#split the data into predictors and target
```

```python

concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

```python
#Let us do a check of the predictors and target data frame
```

```python
predictors.head()
```

```python
target.head()
```

```python
#Let's save the number of predictors to n_cols since we will need this number when building our network.
```

```python
n_cols = predictors.shape[1]
print(n_cols)
```

```python
#import necessary Sci-kit Learn Libraries
```

```python

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#LETS IMPORT KERAS LIBRARIES WHICH WE WILL BE USING 
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
#we will define a function that defines our regression model for us so that so can create our model
```

```python
# define regression model that we need to model our data
```

```python

def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))#one hidden layer
    model.add(Dense(1)) # output layer
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
#lets define a list which we will store the  MSE of 50 iterations
MSE_A =[] # for part A
```

```python
#LETS SPLIT THE DATA INTO TRAINING AND TESTING SETS USING SKICIT LEARN LIBRARIES AND MODEL IT FOR  ITERATIONS
```

```python
for i in range(50):
    predictors_train, predictors_test, target_train, target_test = train_test_split(predictors, target, test_size=0.3, random_state=i)
    # build the model
    model = regression_model()# calling the function
    model.fit(predictors_train, target_train, epochs=50, verbose=2) # train the model on training data with 50 epochs
    target_predicted= model.predict(predictors_test)# testing on the test set
    msea=mean_squared_error(target_test,target_predicted) # calculating the mean squared error 
    MSE_A.append(msea)   # appending to our list MSE_A
```

```python
#CALCULATE THE MEAN AND Standard Deviation OF THE MSE OF THE ABOVE PART A
```

```python
mean_MSE_A= np.mean(MSE_A)
sd_MSE_A=np.std(MSE_A)
print("MEAN OF THE MSE IN PART A is :",mean_MSE_A)
print("STANDARD DEVIATION OF THE MSE IN PART A is :",sd_MSE_A)
```

```python
#PART B OF THE ASSGINMENT using normalised version of data
```

```python
#normalise the data using mean and standard deviation
predictors_norm = (predictors - predictors.mean()) / predictors.std() # normalised data
predictors_norm.head()
```

```python
for i in range(50):
    predictors_norm_train, predictors_norm_test, target_train, target_test = train_test_split(predictors_norm, target, test_size=0.3, random_state=i)
    # build the model
    model = regression_model() # calling our function
    model.fit(predictors_train, target_train, epochs=50, verbose=2) # train the model on training data with 50 epochs
    target_predicted= model.predict(predictors_test) # testing on the test set
    mseb=mean_squared_error(target_test,target_predicted) # calculating the mean squared error
    MSE_B.append(mseb)   # appending to out list MSE_B
```

```python
#CALCULATE THE MEAN AND Standard Deviation OF THE MSE OF THE ABOVE PART B
```

```python
mean_MSE_B= np.mean(MSE_B)
sd_MSE_B=np.std(MSE_B)
print("MEAN OF THE MSE IN PART B is :",mean_MSE_B)
print("STANDARD DEVIATION OF THE MSE IN PART B is :",sd_MSE_B)
```

```python
#PART D 
#new regression model with increased number of hidden layers i.e,3
```

```python
#LETS DEFINE NEW FUNCTION WHICH DEFINES OUR REGRESSION MODEL AND WE CAN CONVENIENTLY BUILD OUR MODEL
```

```python
# define regression model
def new_regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))# 1st hidden layer
    model.add(Dense(10, activation='relu')) # 2nd hidden layer
    model.add(Dense(10, activation='relu')) # 3rd hidden layer
    model.add(Dense(1)) # output layer
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
#lets define a list which we will store the  MSE of 50 iterations
MSE_D =[] # for part D
```

```python
#LETS SPLIT THE DATA INTO TRAINING AND TESTING SETS USING SKICIT LEARN LIBRARIES AND MODEL IT FOR  ITERATIONS
```

```python
for i in range(50):
    predictors_norm_train, predictors_norm_test, target_train, target_test = train_test_split(predictors_norm, target, test_size=0.3, random_state=i)
    # build the model
    model = new_regression_model() # calling the function
    model.fit(predictors_train, target_train, epochs=50, verbose=2) # train the model on training data with 50 epochs
    target_predicted= model.predict(predictors_test) # testing on the test set
    msed=mean_squared_error(target_test,target_predicted) # calculating the mean square error
    MSE_D.append(msed)   # appending to our list MSE_D
```

```python
#CALCULATE MEAN AND STANDARD DEVIATION OF THE ABOVE PART D
```

```python
mean_MSE_D= np.mean(MSE_D)
sd_MSE_D=np.std(MSE_D)
print("MEAN OF THE MSE IN PART D is :",mean_MSE_D)
print("STANDARD DEVIATION OF THE MSE IN PART D is :",sd_MSE_D)
```

```python
#DIFFERENCE IN MEAN OF MSE in Step D and Step B
```

```python
if(mean_MSE_D > mean_MSE_B):
    print("Mean of MSEs in STEP D is greater than Mean of MSEs in STEP B by : ",mean_MSE_D-mean_MSE_B)
else :
     print("Mean of MSEs in STEP B is greater than Mean of MSEs in STEP D by : ",mean_MSE_B-mean_MSE_D)
```
